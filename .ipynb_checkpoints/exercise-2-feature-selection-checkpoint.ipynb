```python
from sklearn import datasets, preprocessing, feature_selection
from sklearn.datasets import make_classification 
from itertools import compress

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Dataset

```python
x, y = make_classification (
    n_samples = 1000,
    n_features = 20,
    n_informative = 5,
    n_redundant = 5,
    n_repeated = 5,
    shuffle=False
)
```

```python
dataset = pd.DataFrame(x)
dataset.columns = [
    "informative_A", "informative_B", "informative_C", "informative_D", "informative_E",
    "redundant_A", "redundant_B", "redundant_C", "redundant_D", "redundant_E",
    "repeated_A", "repeated_B", "repeated_C", "repeated_D", "repeated_E",
    "unknown_A", "unknown_B", "unknown_C", "unknown_D", "unknown_E",
]
# dataset['label'] = y

dataset.head
```

# Feature selection


```python
def get_selected_data(selector, dataset):
    cols_idxs = selector.get_support(indices=True)
    return dataset.iloc[:, cols_idxs]
```

## Low variance

```python
from sklearn.feature_selection import VarianceThreshold

varThreshold = VarianceThreshold(threshold=(1.5))
dataset_removed_low_var = varThreshold.fit_transform(dataset)

dataset_removed_low_var = get_selected_data(varThreshold, dataset)
dataset_removed_low_var

# varThreshold.variances_
# varThreshold.get_support()
```

## Linear regression

```python
from sklearn.feature_selection import SelectKBest, f_regression, mutual_info_regression, mutual_info_classif

selectKBest = SelectKBest(f_regression, k=5)
selectKBest.fit(dataset_removed_low_var, y)

dataset_linear_reg = get_selected_data(selectKBest, dataset_removed_low_var)
dataset_linear_reg
```

```python
selectKBest = SelectKBest(mutual_info_classif, k=5)
selectKBest.fit(dataset_removed_low_var, y)

dataset_linear_reg_2 = get_selected_data(selectKBest, dataset_removed_low_var)
dataset_linear_reg_2
```

```python
selectKBest = SelectKBest(mutual_info_regression, k=5)
selectKBest.fit(dataset_removed_low_var, y)

dataset_linear_reg_2 = get_selected_data(selectKBest, dataset_removed_low_var)
dataset_linear_reg_2
```

```python
for count, score in enumerate(selectKBest.scores_):
    print(str(count) + ": ", score)
```
